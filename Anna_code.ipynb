```python
minimal_1 = trays['minimal_1']
minimal_2 = trays['minimal_2']
minimal_3 = trays['minimal_3']

baseline_1 = trays['baseline_1']
baseline_2 = trays['baseline_2']
baseline_3 = trays['baseline_3']

maximal_1 = trays['maximal_1']
maximal_2 = trays['maximal_2']
maximal_3 = trays['maximal_3']

efflux_1 = trays['efflux_1']
efflux_2 = trays['efflux_2']
efflux_3 = trays['efflux_3']
```

```python
minimal_avg = (minimal_1 + minimal_2 + minimal_3)/3 

baseline_avg = (baseline_1 + baseline_2 + baseline_3)/3 

maximal_avg = (maximal_1 + maximal_2 + maximal_3)/3 

efflux_avg  = (efflux_1 + efflux_2 + efflux_3)/3
```

```python
# Calculate reference 

Ref = maximal_avg-minimal_avg 

Ratio_baseline =  np.abs((baseline_avg-Ref)/Ref)

Ratio_efflux = np.abs((efflux_avg-Ref)/Ref)

t = np.linspace(1, 30, len(Ref))
# t = np.arang
```

```python
plt.plot(t, Ratio_baseline, marker='^', markersize=7, linestyle='-', label='Baseline')
plt.plot(t, Ratio_efflux, marker='s', markersize=5, linestyle='-', label='Efflux')
plt.xlabel('Time')
plt.ylabel('Relative Flourescence')
plt.title('Efflux of ATVR by M. abscessus') 
plt.legend()
plt.show()
```
